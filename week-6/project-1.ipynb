```python
#updated version of code
import tkinter as tk
from tkinter import messagebox

def calculate_bill(name, size, location):
    if location.lower() == 'ibeju-lekki':
        if size >= 10:
            return f'{name}, your bill is N5,000.'
        else:
            return f'{name}, your bill is N3,500.'
    elif location.lower() == 'epe':
        if size >= 10:
            return f'{name}, your bill is N10,000.'
        else:
            return f'{name}, your bill is N5,000.'
    else:
        return "INVALID LOCATION"

def submit():
    name = name_entry.get()
    size = int(size_entry.get())
    location = location_entry.get()
    bill_message = calculate_bill(name, size, location)
    messagebox.showinfo("Bill", bill_message)

root = tk.Tk()
root.title('SIMI LOGISTICS BY ADEKUNLE GOLD')
root.geometry('500x200')

name_label = tk.Label(root, text='Username')
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

size_label = tk.Label(root, text='Size(in kg)')
size_label.pack()
size_entry = tk.Entry(root)
size_entry.pack()

location_label = tk.Label(root, text='Location')
location_label.pack()
location_entry = tk.Entry(root)
location_entry.pack()

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

submit_button.config(fg='black', bg='pink')

root.mainloop()
```
