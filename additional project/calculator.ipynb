```python
import tkinter as tk
from tkinter import END

def solve(number):
    a=result.get()
    result.delete(0,END)
    result.insert(0,str(a) + str(number))
    return;

def clear():
    result.delete(0,END)
    return;

def add():
    global operation
    operation='+'
    first = result.get()
    global f 	                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                
    f = int(first)
    result.delete(0,END)
def minus():
    global operation
    operation='-'
    first = result.get()
    global f 
    f = int(first)
    result.delete(0,END)
def multiply():
    global operation
    operation='x'
    first = result.get()
    global f 
    f = int(first)
    result.delete(0,END)
def divide():
    global operation
    operation='/'
    first = result.get()
    global f 
    f = int(first)
    result.delete(0,END)
def equal():
    second=result.get()
    s = int(second)
    result.delete(0,END)
    if operation == '+': 
        result.insert(0,float(f) + float(s))
    if operation == '-': 
        result.insert(0,float(f) - float(s))
    if operation == 'x': 
        result.insert(0,float(f) * float(s))
    if operation == '/': 
        if s==0:
            result.insert(0,'ERROR')
        else:
            result.insert(0,float(f) / float(s))
    
root = tk.Tk()
root.title('Simple Calculator')
root.geometry('288x320')
root.config(bg='pink')
button1=tk.Button(root,text='1',bg='white',fg='black',height=3, width=8,command=lambda:solve(1))
button1.place(x=5,y=55)
button2=tk.Button(root,text='2',bg='white',fg='black',height=3,width=8,command=lambda:solve(2))
button2.place(x=75,y=55)
button3=tk.Button(root,text='3',bg='white',fg='black',height=3,width=8,command=lambda:solve(3))
button3.place(x=145,y=55)
button4=tk.Button(root,text='4',bg='white',fg='black',height=3,width=8,command=lambda:solve(4))
button4.place(x=5,y=120)
button5=tk.Button(root,text='5',bg='white',fg='black',height=3,width=8,command=lambda:solve(5))
button5.place(x=75,y=120)
button6=tk.Button(root,text='6',bg='white',fg='black',height=3,width=8,command=lambda:solve(6))
button6.place(x=145,y=120)
button7=tk.Button(root,text='7',bg='white',fg='black',height=3,width=8,command=lambda:solve(7))
button7.place(x=5,y=185)
button8=tk.Button(root,text='8',bg='white',fg='black',height=3,width=8,command=lambda:solve(8))
button8.place(x=75,y=185)
button9=tk.Button(root,text='9',bg='white',fg='black',height=3,width=8,command=lambda:solve(9))
button9.place(x=145,y=185)
button0=tk.Button(root,text='0',bg='white',fg='black',height=3,width=8,command=lambda:solve(0))
button0.place(x=75,y=250)
buttonx=tk.Button(root,text='x',bg='red',fg='black',height=3,width=8,command=multiply)
buttonx.place(x=215,y=120)
buttonplus=tk.Button(root,text='+',bg='red',fg='black',height=3,width=8,command=add)
buttonplus.place(x=215,y=55)
buttonm=tk.Button(root,text='-',bg='red',fg='black',height=3,width=8,command=minus)
buttonm.place(x=215,y=185)
buttond=tk.Button(root,text='รท',bg='red',fg='black',height=3,width=8,command=divide)
buttond.place(x=145,y=250)
buttonc=tk.Button(root,text='CLR',bg='red',fg='black',height=3,width=8,command=clear)
buttonc.place(x=5,y=250)
buttone=tk.Button(root,text='=',bg='red',fg='black',height=3,width=8,command=equal)
buttone.place(x=215,y=250)

result = tk.Entry(root,font=('Ariel 15'))
result.place(x=35,y=15)

root.mainloop()
```
